```python
# 12. Elabore um algoritmo que leia um número de entrada que indicará a quantidade de
# registros a serem lidos (N). Em seguida algoritmo deve solicitar o nome e idade de N
# pessoas e ao final apresentar o nome da pessoa mais velha.

n = int(input("Digite a quantidade de pessoas: "))

mais_velho_nome = ""
mais_velho_idade = -1
contador = 0

while contador < n:
    nome = input(f"Digite o nome da {contador+1}ª pessoa: ")
    idade = int(input(f"Digite a idade de {nome}: "))

    if idade > mais_velho_idade:  # atualiza se for mais velho
        mais_velho_nome = nome
        mais_velho_idade = idade

    contador += 1

print("\nA pessoa mais velha é:", mais_velho_nome, "com", mais_velho_idade, "anos.")
```

```python
# 13. Elabore um algoritmo que leia um número de entrada que indicará a quantidade de
# registros a serem lidos (N). Em seguida algoritmo deve solicitar o sexo (M/F) e idade de N
# pessoas e ao final apresentar a média de idade de ambos os gêneros catalogados.

```

```python
n = int(input("Digite a quantidade de pessoas: "))

soma_m = 0
qtd_m = 0
soma_f = 0
qtd_f = 0

contador = 0

while contador < n:
    sexo = input(f"Digite o sexo da {contador+1}ª pessoa (M/F): ").strip().upper()
    idade = int(input("Digite a idade: "))

    if sexo == "M":
        soma_m += idade
        qtd_m += 1
    elif sexo == "F":
        soma_f += idade
        qtd_f += 1
    else:
        print("⚠ Sexo inválido, registro ignorado.")

    contador += 1

# cálculo das médias
media_m = soma_m / qtd_m if qtd_m > 0 else 0
media_f = soma_f / qtd_f if qtd_f > 0 else 0

print("\nMédia de idade dos homens:", media_m)
print("Média de idade das mulheres:", media_f)
```

```python
# 14. Elabore um algoritmo que solicite ao usuário 10 números reais e ao final apresente o
# maior e o menor deles.

contador = 0
maior = None
menor = None

while contador < 10:
    num = float(input(f"Digite o {contador+1}º número: "))

    if maior is None or num > maior:
        maior = num
    if menor is None or num < menor:
        menor = num

    contador += 1

print("\nO maior número digitado foi:", maior)
print("O menor número digitado foi:", menor)
```

```python
# 15. Elabore um algoritmo que solicite N números reais e quando o usuário informar o valor
# nulo 0 (zero) o programa ordene e mostre todos os números informados de forma crescente.


numeros = []

while True:
    num = float(input("Digite um número (0 para encerrar): "))
    if num == 0:
        break
    numeros.append(num)

# ordena a lista
numeros.sort()

print("\nNúmeros informados em ordem crescente:")
for n in numeros:
    print(n)
```

```python
# 16. Escreva um programa que vá solicitando as idades dos alunos da sala até que todos
# sejam informados (perguntar ao usuário se deseja informar a idade do próximo aluno). Ao
# final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos têm mais de
# 18 anos, quantos alunos têm até 18 anos, a média aritmética e a mediana.

idades = []

while True:
    idade = int(input("Digite a idade do aluno: "))
    idades.append(idade)

    continuar = input("Deseja informar a idade do próximo aluno? (S/N): ").strip().upper()
    if continuar != "S":
        break

# cálculos
mais_novo = min(idades)
mais_velho = max(idades)
maiores_18 = sum(1 for i in idades if i > 18)
ate_18 = sum(1 for i in idades if i <= 18)
media = sum(idades) / len(idades)

# mediana
idades.sort()
n = len(idades)
if n % 2 == 1:  # ímpar
    mediana = idades[n // 2]
else:  # par
    mediana = (idades[n // 2 - 1] + idades[n // 2]) / 2

# saída
print("\n--- Resultados ---")
print("Idade do mais novo:", mais_novo)
print("Idade do mais velho:", mais_velho)
print("Quantidade de alunos com mais de 18 anos:", maiores_18)
print("Quantidade de alunos com até 18 anos:", ate_18)
print("Média aritmética das idades:", media)
print("Mediana das idades:", mediana)
```
